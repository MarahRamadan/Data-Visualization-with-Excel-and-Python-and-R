```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() #overwrite the matplotlib with style of seaborn, we'll code the graphs using matplot but displayed with seaborn
```

```python
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```python
import io

df_used_cars = pd.read_csv(io.BytesIO(uploaded['bar_chart_data.csv']),encoding= 'unicode_escape')
```

```python
df_used_cars
```

```python
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars['Brand'],
        height = df_used_cars['Cars Listings'])
plt.xticks(rotation=45)
plt.show()
```

```python
#after sns.set()
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars['Brand'],
        height = df_used_cars['Cars Listings'])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars['Brand'],
        height = df_used_cars['Cars Listings'],
        color = 'rgbwymc')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars['Brand'],
        height = df_used_cars['Cars Listings'],
        color = 'midnightblue')
plt.xticks(rotation=45, fontsize=13)
plt.yticks(fontsize=13)
plt.title("Car Listings by Brand", fontsize=16, fontweight='bold')
plt.ylabel("Number of Listing", fontsize=13)
plt.show()
```
